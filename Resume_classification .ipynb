```python
def preprocess(sentence):
    sentence = str(sentence)
    sentence = sentence.lower()
    sentence = sentence.replace('{html}',"")
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url = re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]


    return " ".join(filtered_words)
```

```python
from nltk.tokenize import RegexpTokenizer
data = pd.read_csv('resume.csv')
data['resume_details'] = data.RAW_DETAILS.apply(lambda x: preprocess(x))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Assuming your dataset is in CSV format
data = pd.read_csv('Resume.csv')  # Make sure it contains columns 'resume_details' and 'Category'

# Preprocess text (remove stopwords and vectorize)
stop_words = stopwords.words('english')

def preprocess_text(text):
    # Remove stopwords and tokenize
    words = text.split()
    words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(words)

data['RAW_DETAILS'] = data['RAW_DETAILS'].apply(preprocess_text)

X = data['RAW_DETAILS']
y = data['Category']
```

```python
# TF-IDF Vectorizer to convert resume text into numerical format
vectorizer = TfidfVectorizer()
X_vect = vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vect, y, test_size=0.2, random_state=42)
```

```python
# Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)
log_reg_acc = accuracy_score(y_test, log_reg_pred)
print(f'Logistic Regression Accuracy: {log_reg_acc}')
```

```python
# decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
print(f'Decision Tree Accuracy: {dt_acc}')
```

```python
# SVM
svm = SVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
print(f'SVM Accuracy: {svm_acc}')
```

```python
#Random forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print(f'Random Forest Accuracy: {rf_acc}')
```

```python
# Print the accuracies of all models
print(f'Logistic Regression Accuracy: {log_reg_acc}')
print(f'Decision Tree Accuracy: {dt_acc}')
print(f'SVM Accuracy: {svm_acc}')
print(f'Random Forest Accuracy: {rf_acc}')

# Choose the best model based on accuracy
best_model = None
best_accuracy = max(log_reg_acc, dt_acc, svm_acc, rf_acc)

if best_accuracy == log_reg_acc:
    best_model = log_reg
elif best_accuracy == dt_acc:
    best_model = dt
elif best_accuracy == svm_acc:
    best_model = svm
else:
    best_model = rf

print(f'The best model is: {best_model}')
```
